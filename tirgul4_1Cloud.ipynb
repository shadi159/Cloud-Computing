```python
# -*- coding: utf-8 -*-
"""Cars_DB.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1vX90dk7RUqXPNgH-kMnnTPhtwSaDCRUs
"""

url = 'https://data.gov.il/api/3/action/datastore_search?resource_id=053cea08-09bc-40ec-8f7a-156f0677aff3'

import requests
import pandas as pd
pd.set_option("display.max_columns", None)

response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)

data_df = pd.DataFrame(data['result']['records'])

data_df.head()

# Import necessary libraries
from ipywidgets import widgets, VBox, interact, Output
from IPython.display import display
output_area = Output()

# Create widgets
tozeret_nm_dropdown = widgets.Dropdown(
    options=[''] + sorted(data_df['tozeret_nm'].unique().tolist()),
    description='Tozeret:',
    style={'description_width': 'initial'}
)

kinuy_mishari_dropdown = widgets.Dropdown(
    options=[''],
    description='Kinuy Mishari:',
    style={'description_width': 'initial'}
)

# Function to update the second dropdown based on the selection of the first
def update_kinuy_mishari_options(change):
    if change['new']:  # Check if a valid option is selected
        filtered_values = data_df[data_df['tozeret_nm'] == change['new']]['kinuy_mishari'].unique()
        kinuy_mishari_dropdown.options = [''] + sorted(filtered_values)
    else:
        kinuy_mishari_dropdown.options = ['']

# Function to update the output area with the total records and unique 'ramat_gimur' values
def update_output(change=None):
    output_area.clear_output()
    selected_tozeret = tozeret_nm_dropdown.value
    selected_kinuy = kinuy_mishari_dropdown.value

    if selected_tozeret and selected_kinuy:
        filtered_df = data_df[
            (data_df['tozeret_nm'] == selected_tozeret) &
            (data_df['kinuy_mishari'] == selected_kinuy)
        ]
        total_records = len(filtered_df)
        unique_ramat_gimur = filtered_df['ramat_gimur'].unique()

        with output_area:
            print(f"Total Records: {total_records}")
            print(f"Unique Ramat Gimur: {', '.join(unique_ramat_gimur) if len(unique_ramat_gimur) > 0 else 'None'}")
    else:
        with output_area:
            print("Please select valid options for both dropdowns.")

# Observe changes in both dropdowns
tozeret_nm_dropdown.observe(update_kinuy_mishari_options, names='value')
kinuy_mishari_dropdown.observe(update_output, names='value')

# Display widgets and output area
display(VBox([tozeret_nm_dropdown, kinuy_mishari_dropdown, output_area]))

```

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Create output widgets for each tab
output_stats = widgets.Output()
output_table = widgets.Output()
output_chart_year = widgets.Output()
output_sunburst = widgets.Output()
output_heatmap = widgets.Output()
output_filter = widgets.Output()

# Tab 1: show statistics
def show_statistics():
    with output_stats:
        clear_output()
        display(data_df.describe())
        plt.figure(figsize=(10, 4))
        sns.histplot(data_df['shnat_yitzur'], bins=20, kde=True)
        plt.title('Date Distribution')
        plt.xlabel('Age')
        plt.ylabel('Frequency')
        plt.show()

# Tab 2: Full Data Table
def show_table():
    with output_table:
        clear_output()
        display(data_df)

# Tab 3: Cars per Year Chart
def show_year_chart():
    with output_chart_year:
        clear_output()
        if 'shnat_yitzur' in data_df.columns:
            df_temp = data_df.copy()
            df_temp['shnat_yitzur'] = pd.to_numeric(df_temp['shnat_yitzur'], errors='coerce')
            year_counts = df_temp['shnat_yitzur'].dropna().astype(int).value_counts().sort_index()
            if not year_counts.empty: # Add check for empty data
                fig, ax = plt.subplots(figsize=(12, 6))

                # Create the bar chart
                ax.bar(
                    x=year_counts.index.astype(str), # Convert years to string for clear ticks
                    height=year_counts.values,
                    color='skyblue'
                )
                ax.set_title('mispar_rechev using shnat_yitzur', fontsize=14)
                ax.set_xlabel('shnat_yitzur', fontsize=12)
                ax.set_ylabel('mispar_rechev', fontsize=12)
                plt.xticks(rotation=45, ha='right')
                ax.grid(axis='y', linestyle='--', alpha=0.7)
                plt.tight_layout()
                plt.show()
            else:
                print("No valid year data to plot.")
        else:
            print("there is no 'shnat_yitzur' column in the data")


# Tab 4: Sunburst Chart
def show_sunburst():
      with output_sunburst:
        clear_output()

        if 'tozeret_nm' in data_df.columns and 'kinuy_mishari' in data_df.columns:
            # ---- prepare data ----
            sunburst_data = (
                data_df[['tozeret_nm', 'kinuy_mishari']]
                .dropna()
                .groupby(['tozeret_nm', 'kinuy_mishari'])
                .size()
                .reset_index(name='count')
            )

            # 1) limit number of manufacturers (outer ring)
            top_manufacturers = (
                data_df['tozeret_nm']
                .value_counts()
                .head(6)
                .index
            )
            sunburst_data = sunburst_data[
                sunburst_data['tozeret_nm'].isin(top_manufacturers)
            ]

            # 2) keep only top 3 models per manufacturer (inner ring)
            sunburst_data = (
                sunburst_data
                .sort_values(['tozeret_nm', 'count'], ascending=[True, False])
            )
            sunburst_data['rank'] = sunburst_data.groupby('tozeret_nm')['count'] \
                                                 .rank(method='first', ascending=False)
            sunburst_data = sunburst_data[sunburst_data['rank'] <= 3]

            # ---- outer ring: manufacturers ----
            manuf_counts = sunburst_data.groupby('tozeret_nm')['count'].sum()
            outer_sizes = manuf_counts.values
            outer_labels = manuf_counts.index

            # ---- inner ring: models ----
            inner_data = sunburst_data.copy()

            # 3) label only "big enough" models
            MIN_MODEL_COUNT = 50   # you can play with this number
            inner_labels = inner_data.apply(
                lambda r: r['kinuy_mishari'] if r['count'] >= MIN_MODEL_COUNT else '',
                axis=1
            )

            fig, ax = plt.subplots(figsize=(10, 8))

            # outer ring (manufacturers)
            ax.pie(
                outer_sizes,
                labels=outer_labels,
                radius=1.3,
                startangle=90,
                wedgeprops=dict(width=0.3, edgecolor='white'),
                textprops={'fontsize': 9}
            )

            # inner ring (models)
            ax.pie(
                inner_data['count'],
                labels=inner_labels,
                radius=1.0,
                startangle=90,
                wedgeprops=dict(width=0.3, edgecolor='white'),
                labeldistance=0.7,        # move labels closer to center
                textprops={'fontsize': 7} # smaller font
            )

            ax.set(aspect="equal")
            plt.title("Distribution of vehicles by manufacturer and model")
            plt.tight_layout()
            plt.show()

        else:
            print("The asked columns is not in the data")


# Tab 5: Heatmap
def show_heatmap():
    with output_heatmap:
        clear_output()
        if 'tozeret_nm' in data_df.columns and 'shnat_yitzur' in data_df.columns:
            # Prepare data
            data_df['shnat_yitzur_clean'] = pd.to_numeric(
                data_df['shnat_yitzur'], errors='coerce'
            )
            heatmap_data = data_df[['tozeret_nm', 'shnat_yitzur_clean']].dropna()

            # Get top 15 manufacturers
            top_manufacturers = (
                data_df['tozeret_nm']
                .value_counts()
                .head(15)
                .index
            )
            heatmap_data = heatmap_data[
                heatmap_data['tozeret_nm'].isin(top_manufacturers)
            ]

            # Create pivot table
            pivot = (
                heatmap_data
                .groupby(['tozeret_nm', 'shnat_yitzur_clean'])
                .size()
                .reset_index(name='count')
            )
            pivot_table = pivot.pivot(
                index='tozeret_nm',
                columns='shnat_yitzur_clean',
                values='count'
            ).fillna(0)

            # ----- Matplotlib heatmap -----
            fig, ax = plt.subplots(figsize=(12, 8))

            im = ax.imshow(pivot_table.values, aspect='auto')

            # X axis (years)
            x_labels = pivot_table.columns
            ax.set_xticks(np.arange(len(x_labels)))
            ax.set_xticklabels(x_labels.astype(int), rotation=45, ha='right')

            # Y axis (manufacturers)
            y_labels = pivot_table.index
            ax.set_yticks(np.arange(len(y_labels)))
            ax.set_yticklabels(y_labels)

            ax.set_xlabel("shnat_yitzur")
            ax.set_ylabel("manufacturer")
            ax.set_title("Heatmap for shnat_yitzur mispar_rechev and by manufacturer")

            # Colorbar
            cbar = fig.colorbar(im, ax=ax)
            cbar.set_label("mispar_rechev")

            plt.tight_layout()
            plt.show()
        else:
            print("The asked columns is not in the data")
# Tab 6: Filter by Manufacturer and Model
def create_filter_tab():
    with output_filter:
        clear_output()

        # Get unique values
        manufacturers = sorted(data_df['tozeret_nm'].dropna().unique()) if 'tozeret_nm' in data_df.columns else []

        # Create widgets
        manufacturer_dropdown = widgets.Dropdown(
            options=['choose manufacturer...'] + manufacturers,
            description='manufacturer:',
            style={'description_width': '80px'}
        )

        model_dropdown = widgets.Dropdown(
            options=['choose sug_degem...'],
            description='sug_degem:',
            style={'description_width': '80px'},
            disabled=True
        )

        finish_dropdown = widgets.Dropdown(
            options=['choose ramat_gimur...'],
            description='ramat_gimur:',
            style={'description_width': '80px'},
            disabled=True
        )

        result_output = widgets.Output()

        def on_manufacturer_change(change):
            if change['new'] != 'choose manufacturer...':
                # Filter models for selected manufacturer
                filtered_df = data_df[data_df['tozeret_nm'] == change['new']]
                models = sorted(filtered_df['kinuy_mishari'].dropna().unique()) if 'kinuy_mishari' in filtered_df.columns else []
                model_dropdown.options = ['choose sug_degem...'] + list(models)
                model_dropdown.disabled = False
                finish_dropdown.options = ['choose ramat_gimur	...']
                finish_dropdown.disabled = True

                with result_output:
                    clear_output()
            else:
                model_dropdown.options = ['choose sug_degem...']
                model_dropdown.disabled = True
                finish_dropdown.options = ['choose ramat_gimur...']
                finish_dropdown.disabled = True
                with result_output:
                    clear_output()

        def on_model_change(change):
            if change['new'] != 'choose manufacturer...' and manufacturer_dropdown.value != 'choose manufacturer...':
                # Filter by manufacturer and model
                filtered_df = data_df[
                    (data_df['tozeret_nm'] == manufacturer_dropdown.value) &
                    (data_df['kinuy_mishari'] == change['new'])
                ]

                # Get finish levels if column exists
                if 'ramat_gimur' in filtered_df.columns:
                    finishes = sorted(filtered_df['ramat_gimur'].dropna().unique())
                    finish_dropdown.options = ['everything'] + list(finishes)
                    finish_dropdown.disabled = False

                with result_output:
                    clear_output()

                    if 'ramat_gimur' in filtered_df.columns:
                        finish_counts = filtered_df['ramat_gimur'].value_counts()
                        for finish, count in finish_counts.items():
                            print(f"   • {finish}: {count:,} cars")

                    if 'shnat_yitzur' in filtered_df.columns:
                        years = pd.to_numeric(filtered_df['shnat_yitzur'], errors='coerce').dropna()
                        if len(years) > 0:
                            print(f"   from-{int(years.min())} to {int(years.max())}")
                    display(filtered_df.head(10))
            else:
                finish_dropdown.options = ['choose ramat_gimur...']
                finish_dropdown.disabled = True
                with result_output:
                    clear_output()

        def on_finish_change(change):
            if change['new'] != 'choose ramat_gimur...' and change['new'] != 'everything':
                # Filter by all three criteria
                filtered_df = data_df[
                    (data_df['tozeret_nm'] == manufacturer_dropdown.value) &
                    (data_df['kinuy_mishari'] == model_dropdown.value) &
                    (data_df['ramat_gimur'] == change['new'])
                ]

                with result_output:
                    clear_output()
                    if 'shnat_yitzur' in filtered_df.columns:
                        years = pd.to_numeric(filtered_df['shnat_yitzur'], errors='coerce').dropna()
                        if len(years) > 0:
                            print(f"   from-{int(years.min())} to {int(years.max())}")

                    display(filtered_df)
            elif change['new'] == 'everything':
                on_model_change({'new': model_dropdown.value})

        manufacturer_dropdown.observe(on_manufacturer_change, names='value')
        model_dropdown.observe(on_model_change, names='value')
        finish_dropdown.observe(on_finish_change, names='value')

        display(widgets.VBox([
            widgets.HTML('<h3>Filter by manufacturer, model and trim level</h3>'),
            manufacturer_dropdown,
            model_dropdown,
            finish_dropdown,
            result_output
        ]))

# Initialize all tabs
show_statistics()
show_table()
show_year_chart()
show_sunburst()
show_heatmap()
create_filter_tab()

# Create tab widget
tab = widgets.Tab()
tab.children = [
    output_stats,
    output_table,
    output_chart_year,
    output_sunburst,
    output_heatmap,
    output_filter
]

tab.set_title(0, ' statistics')
tab.set_title(1, ' Data table')
tab.set_title(2, ' Annual graph')
tab.set_title(3, ' Sunburst')
tab.set_title(4, ' Heatmap')
tab.set_title(5, ' Advanced filtering')

# Display the dashboard
display(widgets.HTML('<h1 style="text-align:center;">Vehicle data analysis system</h1>'))
display(tab)
```
