```python
import json
import ipywidgets as widgets
from IPython.display import display

# Read the JSON file from Google Drive
with open('/content/drive/MyDrive/Cloud computing/lab 2/students.json', "r") as fid:
  json_string_from_file = json.load(fid)

data_from_file = json.loads(json_string_from_file)

# Extract first names
first_names = [member_data['First name'] for member_data in data_from_file['Group'].values()]

# Create and display the dropdown
first_name_dropdown = widgets.Dropdown(
    options=first_names,
    description='Select:',
)

display(first_name_dropdown)

# The chosen member will be accessed directly from the dropdown's value in the next cell
```

```python
def display_member_info(change, data_from_file):
  selected_first_name = change['new']

  member_found = False
  for member_key, member_data in data_from_file['Group'].items():
    if member_data['First name'] == selected_first_name:
      print(f"--- Information for {selected_first_name} ---")
      for key, value in member_data.items():
        if isinstance(value, list):
          print(f"{key}:")
          for item in value:
            if isinstance(item, dict):
              for sub_key, sub_value in item.items():
                print(f"  - {sub_key}: {sub_value}")
            else:
              print(f"  - {item}")
        else:
          print(f"{key}: {value}")
      member_found = True
      break

  if not member_found:
    print(f"No member found with the first name: {selected_first_name}")

# Observe changes in the dropdown value and call the display function
first_name_dropdown.observe(lambda change: display_member_info(change, data_from_file), names='value')

# Display initial information for the default selected value
display_member_info({'new': first_name_dropdown.value}, data_from_file)
```
