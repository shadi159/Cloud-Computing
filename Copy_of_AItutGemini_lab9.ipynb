```python
import google.generativeai as genai
import textwrap
from IPython.display import display
from IPython.display import Markdown

# --- SETUP ---
# Replace with your actual API key
genai.configure(api_key="key")

# --- DATA & INDEXING (The "Patterns" Variable) ---
# This matches the Phase 3 requirement to use your index/data
patterns = [
    "Collage emerged in 13th-century Europe, evolving through medieval & Victorian uses.",
    "Pablo Picasso and Georges Braque are credited with popularizing collage in 1912.",
    "The term 'collage' comes from the French verb 'coller', meaning 'to glue'."
]

# --- PHASE 3: PROMPT ENGINEERING ---
# We inject the patterns into the system instruction
context_string = "\n".join(patterns)

system_instruction = f"""
You are a helpful assistant for a project about the History of Art.
Your goal is to answer questions based on the specific knowledge provided below.

Knowledge Base:
{context_string}

If the answer is not in the knowledge base, answer based on general knowledge but mention that it wasn't in the specific records.
"""

# Initialize the model (Using gemini-1.5-flash as 2.5 is likely a typo/beta in the prompt)
model = genai.GenerativeModel(
    model_name='gemini-2.5-flash',
    system_instruction=system_instruction
)

# Start the chat session
chat = model.start_chat(history=[])

# --- HELPER FUNCTION ---
def to_markdown(text):
    text = text.replace('•', ' *')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

# --- MAIN LOOP ---
def run_chat():
    print("--- Chatbot Started (Type 'quit' to exit) ---")

    # 3 Example Questions
    example_questions = [
        "1. When did collage first emerge?",
        "2. Who popularized collage in 1912?",
        "3. What is the origin of the word 'collage'?"
    ]

    print("\nHere are 3 example questions you can ask:")
    for q in example_questions:
        print(q)
    print("\n" + "-"*30 + "\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() in ['quit', 'exit', 'bye']:
            print("Bot: Goodbye!")
            break

        try:
            response = chat.send_message(user_input)
            print(f"Bot: {response.text}")
            print("-" * 20)
        except Exception as e:
            print(f"An error occurred: {e}")

if __name__ == "__main__":
    run_chat()
```
